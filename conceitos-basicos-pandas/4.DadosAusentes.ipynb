# Manipulação de Dados Ausentes

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"A": [1, 2, np.nan], "B": [5, np.nan, np.nan], "C": [1, 2, 3]})
```

```python
df
```

```python
df.dropna(axis=0)
```

```python
df.dropna(
    axis=1,
    thresh=2,
)
```

```python
df.fillna(0)
```

```python
df["A"].fillna(value=df["A"].mean())
```

```python
df.ffill()
```
